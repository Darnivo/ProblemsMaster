```python
import math

total_perimeter = 0  # Initialize the total perimeter sum

total_square_area = 0

# Open the file 'Q2.txt' for reading
with open('Q2.txt', 'r') as file:
    for line in file:
        # Read height and width from each line, strip to remove extra spaces or newlines
        height, width = map(int, line.strip().split())

        # Calculate the area
        area = height * width
        
        # Check if the area is a perfect square
        if math.isqrt(area) ** 2 == area:
            # It's a square, so we ignore this polygon
            # print("SQUARESQUARESQUARE")
            total_square_area += area
            continue
        else:
            # It's not a square, calculate the perimeter
            perimeter = 2 * (height + width)
            total_perimeter += perimeter

        # print(f"Height: {height}, Width: {width}, Area: {area}, Perimeter: {2 * (height + width)}")

# Output the total perimeter of non-square polygons
print(f"Total Perimeter of Non-Square Polygons: {total_perimeter}")
print(f"Total Area of Square Polygons: {total_square_area}")

print (total_perimeter % (total_square_area%100))
```

```python
import random
import math

for _ in range(200):  # Generate 10 random test cases
    # Randomly decide whether to generate a pair that gives a perfect square area
    is_square_area = random.choices([True, False], weights=[0.4, 0.6])[0]
    
    if is_square_area:
        # Generate a perfect square area: height * width = n^2
        n = random.randint(2, 10)  # Start from 2 to avoid 1 being chosen
        perfect_square_area = n * n
        
        # Find factors of perfect_square_area to form height and width, excluding 1
        factors = []
        for i in range(2, int(math.sqrt(perfect_square_area)) + 1):  # Start from 2 to avoid 1
            if perfect_square_area % i == 0 and i != 1 and perfect_square_area // i != 1:
                factors.append((i, perfect_square_area // i))
        
        # Randomly choose one factor pair (height, width) that results in the perfect square area
        height, width = random.choice(factors)
    else:
        # Generate a non-square combination (height * width is not a perfect square)
        height = random.randint(2, 10)  # Exclude 1
        width = random.randint(2, 10)  # Exclude 1
        while math.isqrt(height * width) ** 2 == height * width:  # Ensure it's not a perfect square area
            width = random.randint(2, 10)  # Exclude 1
    
    print(f"{height} {width}")
```
