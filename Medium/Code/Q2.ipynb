```python
def solve_problem(file_path):
    # Read the input from the given file
    with open(file_path, 'r') as file:
        # Read lines and parse them into a list of tuples (the original list)
        data = [tuple(map(int, line.strip().split())) for line in file.readlines()]

    # Part 1: Sorting in ascending order
    # Create a list of tuples with (index, tuple), where index is the original position
    indexed_data = [(i + 1, data[i]) for i in range(len(data))]

    # Sort the tuples based on the tuple values (lexicographically)
    sorted_data_asc = sorted(indexed_data, key=lambda x: x[1])

    # Calculate the weighted sum for ascending order
    weighted_sum_asc = sum(index * (i + 1) for i, (index, _) in enumerate(sorted_data_asc))
    print(f"Part 1 Weighted Sum (ascending): {weighted_sum_asc}")

    # Part 2: Sorting in descending order
    sorted_data_desc = sorted(indexed_data, key=lambda x: x[1], reverse=True)

    # Calculate the weighted sum for descending order
    weighted_sum_desc = sum(index * (i + 1) for i, (index, _) in enumerate(sorted_data_desc))
    print(f"Part 2 Weighted Sum (descending): {weighted_sum_desc}")

# Example usage
solve_problem("Q2.txt")
```
