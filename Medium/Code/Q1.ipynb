```python
classes = []

with open('Q1.txt', 'r') as file:
    for line in file:
        classes.append([int(x) for x in line.strip().split()])
        
# print(classes)

medians = []

for cls in classes:
    sorted_class = sorted(cls)
    
    # print(sorted_class)
    median = sorted_class[len(sorted_class) // 2]
    medians.append(median)
    # print(median)
    
sorted_medians = sorted(medians)
print(sorted_medians)

gap = sorted_medians[-1] - sorted_medians[0]
print(gap)

difference = sum([abs(gap - x) for x in sorted_medians])

print(difference)
```
