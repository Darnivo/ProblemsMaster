```python
inputs = []
# inputs.append("1234 5678 9")
# inputs.append("1770 0131 01")
# inputs.append("1770 0131")

with open("Q1.txt") as file:
    for line in file:
        inputs.append(line.strip())

valid_sum = 0

for input in inputs:
    
    print(input,end=" ")
    
    input = input.replace(" ", "")
    
    sum = 0
    
    for i, char in enumerate(input):
        digit = int(char)
        if i % 2 == 1:
            digit = (digit * 2) % 10
        
        sum += digit
    
    print("\nsum >",sum)
    
    if sum % 10 == 0:
        valid_sum += 1
        print("Valid")
    else:
        print("Invalid")
        pass
        
    
    

print(valid_sum)
```
